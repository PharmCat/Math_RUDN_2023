**Числовые множества**

- Множество натуральных чисел $N = \{0, 1, 2, ..., n, ...\}$. Множество натуральных чисел включает в себя все целые неотрицательные (существует подход, где 0 не входит в множество натуральных чисел). 


- Множество целых чисел $Z = \{0, \pm 1, \pm 2, ..., \pm n, ...\}$. В это числовое множество входят все целые числа больше и меньше нуля, а так же ноль. Например: -15, 0, 139.


- Множество рациональных чисел $Q = \{\frac{m}{n}|m, n \in Z, n \neq 0 \}$. Рациональные числа, вообще говоря, представляют собой множество дробей. Пример чисел входящих в рациональное множество: 3/5, 9/7, 1/2. 


- Множество вещественных чисел $R = \{a_1a_2...a_n,b_1b_1...\}$, где $a_1a_2...a_n$ – конечная последовательность цифр целой части числа, принадлежащего множеству вещественных чисел, а $b_1b_2...$ – бесконечная последовательность чисел, стоящих в дробной части вещественного числа. Иными словами, это объединение всех рациональных и всех иррациональных чисел. Например: 1, -2,829 , $\pi$, $e$.

-  Множество комплексных чисел обычно обозначается символом $C$. числа вида ${\displaystyle a+bi,}$ где $a,b$ — вещественные числа, $i$ — мнимая единица, то есть число, для которого выполняется равенство: ${\displaystyle i^{2}=-1.}$



**Функция**

Зависимость одной переменной от другой называется функциональной зависимостью. Зависимость переменной $y$ от переменной $x$ называется функцией, если каждому значению $x$ соответствует единственное значение $y$.

Обозначение: $y=f(x)$

Переменную $x$ называют **независимой** переменной или аргументом, а переменную $y$ - зависимой. Говорят, что $y$ является функцией от $x$. Значение $y$, соответствующее заданному значению $x$, называют значением функции.

Все значения, которые принимает $x$, образуют **область определения функции**; все значения, которые принимает $y$, образуют **множество значений функции**.

**Область определения**  — множество, на котором задаётся функция. В каждой точке этого множества значение функции должно быть определено.


Более формально, если задана функция $f$, которая отображает множество $X$ в $Y$, то есть: $f\colon X\to Y$, то множество $X$ называется областью определения или областью задания функции $f$ и обозначается $D(f)$ или $\mathrm {dom} \,f$. 

**Непрерывной** областью определения называется  функция, которая меняется без мгновенных «скачков» (называемых разрывами), то есть такая, малые изменения аргумента ($x$) которой приводят к малым изменениям значения функции ($y$). График непрерывной функции является непрерывной линией. Геометрический смысл области определения в том, что график будет существовать только в ней.

Так, если дана функция, содержащая дробь $\frac{1}{\alpha(x)}$, то в ее область определения не будут входить те значения аргумента, которые обращают знаменатель в ноль.

Например: $f(x) = \frac{x+2}{x^2 - 3}$ - найти область определения. 

Знаменатель должен быть ненулевым.

$
x^2-3 = 0 \\
x^2 = 3\\
x = \pm \sqrt{3}
$

То есть в область определения будут входить все точки, кроме этих, т.к. при них функция не существует. Ответ: $D(f) = R\setminus \{-\sqrt{3}, \sqrt{3}\}$

Если задана функция с квадратным корнем $\sqrt{\alpha (x)}$, то она будет определена только при тех значениях $x$, когда подкоренное выражение неотрицательно: $\alpha (x) \geq 0$. Если корень расположился в знаменателе $\frac{1}{\sqrt{\alpha (x)}}$, то условие очевидным образом ужесточается: $\alpha(x)>0$. Аналогичные выкладки справедливы для любого корня положительной чётной степени.

Если некоторая функция содержит логарифм $\ln{\alpha(x)}$, то в её область определения должны входить только те значения $x$, которые удовлетворяют неравенству $\alpha(x)>0$. Если логарифм находится в знаменателе: $\frac{1}{\ln{\alpha (x)}}$, то дополнительно накладывается условие $\alpha(x) \neq 1$ (так как $\ln{1}=0$).

**Формальное определение непрерывности:**

Функция $f$ непрерывна в точке $x_{0}$, предельной для множества $D$, если $f$ имеет предел в точке $x_{0}$, и этот предел совпадает со значением функции $f(x_{0})$:

$\lim _{x\to x_{0}}f(x)=f(x_{0})$

**Элементарная функция:** 

Элементaрные функции — функции, которые можно получить с помощью конечного числа арифметических действий из следующих основных элементарных функций:

- линейная функция типа: $y = ax + b$

- степенная функция с любым действительным показателем ($ y = x^2, y = (x-1)^3$);


- показательная ($y = a^x$) и логарифмическая функции ($y = lnx$);


- тригонометрические и обратные тригонометрические функции ($y=\sin{x}$).


Каждую элементарную функцию можно задать формулой, то есть набором конечного числа символов, соответствующих используемым операциям. Все элементарные функции непрерывны на своей области определения.

**Обратная функция:** функция, обращающая зависимость, выражаемую данной функцией. Например, если функция от $x$ даёт $y$, то обратная ей функция от $y$ даёт $x$. Обратная функция функции $f$ обычно обозначается $f^{-1}$, иногда также используется обозначение $f^{\mathrm{inv}}$.

Функция, имеющая обратную, называется обратимой. 

Функция $g:Y\to X$ называется обратной к функции $f:X\to Y$, если выполнены следующие тождества:

- $f(g(y))=y$ для всех $y\in Y$;

- $g(f(x))=x$ для всех $x\in X$.


**Предел функции**

Функция $f(x)$ имеет предел $A$ в точке $x_{0}$, если для всех значений $x$, достаточно близких к $x_{0}$, значение $f(x)$ близко к $A$.

Число $b$ называется пределом функции $f(x)$ в точке $a$, если $\forall \varepsilon >0$ ∃ $\delta >0$, такое что $\forall x,0<|x-a|<\delta $ выполняется ${\displaystyle |f(x)-b|<\varepsilon }$.

Т.е. Если мы выберем допустимое отклонение от числа $b$  -  $\varepsilon$, то всегда найдется такое значение $\delta$, что для всех $x$, удаленных от точки $a$ не боле чем на $\delta$ ($|x - a| < \delta$), отклонение значения функции от числа $b$ не будет превосходить $\varepsilon$ (т.е. $|f(x) - b| < \varepsilon$).

$
\lim\limits_{x \to 1}\frac{2x^2-3x-5}{x+1}
$ 

В данном случае "икс стремится к единице" означает, что икс будет последовательно принимать значения, которые бесконечно близко приближаются к единице и практически с ней совпадают. Чтобы решить такой предел, вместо икса достаточно поставить единицу: 

$
\lim\limits_{x \to 1}\frac{2x^2-3x-5}{x+1} = \frac{2*1^2-3*1-5}{1+1} = \frac{-6}{2} = -3
$ 

Когда в пределе $x$ стремится к бесконечности, это означает, что он неограниченно возрастает до бесконечности, мы также можем просто поставить вместо $x$ бесконечность: 

$\lim\limits_{x \to \infty} (1 - x) = 1 - \infty = - \infty$



Если, подставляя бесконечность в предел, получается выражение вида $\frac{1}{\infty}$, то есть любое число делить на бесконечность, мы будем получать 0. 

$\lim\limits_{x \to \infty} \frac{1}{x-100} = 0 \lim\limits_{x \to \infty} \frac{3}{x^3} = 0 \lim\limits_{x \to \infty} \frac{5}{4^x} = 0$

В случае $\lim\limits_{x \to 0} \frac{1}{x} = \infty$ можно говорить что предел стремится к бесконечности или не существует (DNE). 



**Свойства:**

- Если предел последовательности существует, то он единственный.


- ${\,c={\rm {const}}.}$


- $\lim(f(x) + g(x))=\lim f(x)+\lim g(x)$ (если оба предела существуют)


- $\lim(qf(x))=q\lim f(x) $


- $\lim(f(x)g(x))=\lim f(x)\lim g(x)$ (если оба предела существуют)


- $\lim(f(x)/g(x))=\lim f(x)/\lim g(x)$ (если оба предела существуют и знаменатель правой части не ноль)


**Неопределенность**

Встречаются пределы с неопределенностью вида $\frac{\infty}{\infty}$, тогда общепринятым алгоритмом считается вынесение старшей степени:

$\lim\limits_{x \to \infty}\frac{2x^2-3x-5}{1+x+3x^2} $ - здесь в числителе и знаменателе старшая степень равняется двум. 

Тогда разделим почленно каждый элемент на $x^2$: $\lim\limits_{x \to \infty}\frac{2x^2/x^2-3x/x^2-5/x^2}{1/x^2+x/x^2+3x^2/x^2} = \lim\limits_{x \to \infty}\frac{2-(3/x)^{\to 0}-(5/x^2)^{\to 0}}{(1/x^2)^{\to 0}+(1/x)^{\to 0}+3}$


Если в числителе и знаменателе находятся многочлены, и имеется неопределенности вида $\frac{0}{0}$, то для ее раскрытия нужно разложить числитель и знаменатель на множители*, либо домножить на сопряженное выражение. (Сопряженное выражение - это такое выражение, на которое надо домножить исходное выражение, чтобы получить разность квадратов)

$\lim\limits_{x \to 1} = \frac{x^2 + 2x - 3}{x^2 - 3x + 2} = \frac{(x-1)(x+3)}{(x-1)(x-2)} = \frac{x+3}{x-2} = \frac{4}{-1}$

$\lim\limits_{x \to -1} = \frac{x + 1}{x + \frac{1}{x} + 2} = \frac{x(x + 1)}{x^2 + 1 + 2x} = \frac{x(x + 1)}{(x+1)^2} = \frac{x}{x+1}$

*см. разложение многочлена на множители

Также неопределенности $\frac{0}{0}$ и $\frac{\infty}{\infty}$ могут быть решены с помощью правила Лопиталя: 

$\lim\limits_{x \to a}\frac{f(x)}{g(x)} = \lim\limits_{x \to a}\frac{f'(x)}{g'(x)}$

если:

* $\lim\limits_{x \to a} f(x) = \lim\limits_{x \to a} g(x) ∈ \{0 , \infty\}$


* $\lim\limits_{x \to a} g'(x) ≠ 0 $


* существует $  \lim\limits_{x \to a}\frac{f'(x)}{g'(x)}$


$\lim\limits_{x \to 1} = \frac{x^2 + 2x - 3}{x^2 - 3x + 2} = \frac{2x + 2}{2x-3} = \frac{4}{-1}$



**Предел последовательности**

Пределом последовательности называют объект, к которому члены последовательности в некотором смысле стремятся или приближаются с ростом порядкового номера.

Число $a$ a называется пределом последовательности $ x_{1},x_{2},...,x_{n},...$, если ${\displaystyle \forall {\text{  }}\varepsilon >0{\text{,  }}\exists {\text{  }}N(\varepsilon ){\text{,   }}\forall {\text{  }}n>N(\varepsilon ){\text{  }}{\text{  }}|x_{n}-a|<\varepsilon }$.

Предел последовательности обозначается ${\displaystyle \lim _{n\to +\infty }x_{n}}$. 



**Непрерывность функции** - уже было сказано, что это функция, которая меняется без мгновенных «скачков» (называемых разрывами), то есть такая, малые изменения аргумента которой приводят к малым изменениям значения функции. График непрерывной функции является непрерывной линией.

Можно рассмотреть для наглядности график некоторой функции $y=f(x)$



В данном случае  функция определена на всей числовой прямой, то есть для каждого значения «икс» существует своё значение «игрека» $y=f(x)$. Одна точка выколота, так как по определению функции, значению аргумента должно соответствовать единственное значение функции. Однако эта функция не является непрерывной на $R$, так как в точке $x=k$ она терпит разрыв.

Важно понимать определение непрерывности функции в точке. Для этого нам потребуются односторонние пределы, которые описывают приближение к точке с правой или с левой стороны. 

Так, левосторонним пределом будет называться $\lim\limits_{x \to k^-}f(x)=m$, где $k^-$ означает приближение к точке $k$ слева, и аналогично правосторонний $\lim\limits_{x \to k^+}f(x)=m$. Если односторонние пределы конечны и равны, то существует общий предел $\lim\limits_{x \to k}f(x)=m$, который и является пределом функции. Тогда функция непрерывна в точке $k$, если предел функции в данной точке равен значению функции в этой точке: $\lim\limits_{x \to k}f(x)=f(k)$.

То есть, непрерывность в точке гарантируется тремя условиями: 

- функция определена в точке
- существуют и равны друг другу односторонние пределы
- предел функции в данной точке должен быть равен значению функции в этой точке

В общем случае функция проверяется на непрерывность этими тремя условиями.

Если нарушено хотя бы одно из описанных условий, то функция теряет свойство непрерывности в точке $k$. Из этого вытекает определение точек разрыва. Если в точке  нарушено условие непрерывности и односторонние пределы конечны, то она называется точкой разрыва первого рода. При этом если пределы не равны, то разность пределов будет называться скачком функции. Если разность пределов равна нулю, то эта точка будет называться точкой устранимого разрыва. Примером будет служить $f(x) = \frac{\sin{x}}{x}$. Данная функция непрерывна на всей числовой прямой, кроме точки $x=0$, где она не определена и в которой функция терпит устранимый разрыв, т.е. устраним разрыв если перепишем функцию в систему: 

$
f(x) = \begin{cases}
\frac{\sin{x}}{x}, \; если \; x \neq 0\\
1, \; если \; x = 0
\end{cases}
$. 

Разрыв может не быть устранимым в случае, если график имеет скачки. 

Если лево-, правосторонний, или оба равны бесконечности, тогда в точке наблюдается разрыв второго рода.  Примером послужит любая гипербола, у которой лево- и правосторонние пределы равны бесконечности. В данном случае координатная прямая будет выступать в роли ассимптоты, которую график никогда не пересечет.


**Теорема о промежуточном значении** - если непрерывная на отрезке функция принимает значения  $a$ и $b$, то она принимает все значения между ними. 

```julia
fun(x) = sin(x)/x
plot(fun, xlims = (-30, 30), ylims = (-1, 2))
```

**Замечательные пределы**

**Первый замечательный предел** имеет вид: $\lim\limits_{\alpha \to 0}\frac{\sin{\alpha}}{\alpha}=1$.
Если поменять числитель и знаменатель местами, смысл останется таким же: $\lim\limits_{\alpha \to 0}\frac{\alpha}{\sin{\alpha}}=1$

В качестве $\alpha$ может передаваться любая элементарная или сложная функция: 


$
\lim\limits_{x \to 0} \frac{\sin(\arctan (x))}{\arctan (x)} = 1
$


$
\lim\limits_{x \to 1} \frac{x^3 - 5x^2 + 3}{\sin{x^3 - 5x^2 + 3}} = 1
$

Решить предел: $\lim\limits_{x \to 0} \frac{5x^2}{\sin^2{\frac{x}{2}}}$

Видим, что икс стремится к нулю, а в знаменателе есть синус. Подставляем ноль и получаем неопределенность вида $\frac{0}{0}$. Теперь можем использовать первый замечательный предел. Расскладываем на множители: $\lim\limits_{x \to 0} \frac{5x*x}{\sin{\frac{x}{2}}\sin{\frac{x}{2}}}$. Исскуственно восстанавливаем первый замечательный предел - для этого делим иксы в числителе на два и умножаем на два: $\lim\limits_{x \to 0} \frac{5x*x}{\sin{\frac{x}{2}}\sin{\frac{x}{2}}} = \lim\limits_{x \to 0} \frac{5*2*2*x/2*x/2}{\sin{\frac{x}{2}}\sin{\frac{x}{2}}} = 5 * 2 * 2 = 20$


**Второй замечательный предел**: $\lim\limits_{\alpha \to \infty}(1 + \frac{1}{\alpha})^{\alpha}=e$, неопределеность вида $1^{\infty}$. 

*Найти предел $\lim\limits_{x \to \infty}(\frac{x-2}{x+1})^{2x+3}$*.

В данном случае получаем неопределенность вида $(\frac{\infty}{\infty})^{\infty}$, а нам нужно $1^{\infty}$.

Для этого преобразуем основание степени. $\lim\limits_{x \to \infty}(\frac{x-2}{x+1})^{2x+3} =
\lim\limits_{x \to \infty}( \frac{x+1-3}{x+1})^{2x+3} = \lim\limits_{x \to \infty}( \frac{x+1}{x+1} + \frac{-3}{x+1})^{2x+3}
\lim\limits_{x \to \infty}(1 + \frac{-3}{x+1})^{2x+3}$.

Теперь "переворачиваем" дробь:

$=\lim\limits_{x \to \infty}\bigg(1 +\frac{1}{\frac{x+1}{-3}}\bigg)^{2x+3}$ 

Имеем необходимую нам неопределенность $1^{\infty}$, а скобка приняла вид  $\Big(1 + \frac{1}{\alpha}\Big)$.  

Производим манипуляции с степенями и применяем второй замечательный предел. 

$ =\lim\limits_{x \to \infty}\bigg(1 +\frac{1}{\frac{x+1}{-3}}\bigg)^{2x+3} = 
 \lim\limits_{x \to \infty}\Bigg(\bigg(1 +\frac{1}{\frac{x+1}{-3}}\bigg)^{\frac{x+1}{-3}}\Bigg)^{\frac{-3}{x+1}*2x+3} = e^{-3\lim\limits_{x \to \infty}\frac{2x+3}{x+1}}$ 

Снова неопределенность. Выносим старшую степень: делим числитель и знаменатель на $x$:

$
 = e^{-3\lim\limits_{x \to \infty}\frac{2x+3}{x+1}}=e^{\frac{\infty}{\infty}} = e^{-3\lim\limits_{x \to \infty}\frac{2+3/x}{1+1/x}}= e^{-3*2} = e^{-6}$ 

**Возрастание и убывание функций, максимумы и минимумы, наибольшие и наименьшие значения функций**

Промежутки монотонности, или интервалы возрастания/убывания находятся с помощью производных, которые отражают скорость изменения функции в точке. Функция возрастает, то есть каждое следующее её значение больше предыдущего, убывает  – каждое следующее значение меньше предыдущего. 


Нам дана дифференцируемая функция на некотором интервале $y = f(x)$. Тогда: 

- если производная $f'(x) > 0 $ на интервале, то функция $f(x)$ возрастает на данном интервале;

- если производная $f'(x) < 0 $ на интервале, то функция $f(x)$ убывает на данном интервале.

Если производная в точке будет равна нулю, то точка будет называться **стационарной**.

Если производная в точке будет равна нулю или не существует, то точка будет называться **критической**.

**Локальные/глобальные экстремумы (минимумы и максимумы)**


**Точкой глобального минимума** называется такая точка $a$, что для лобого $x$ из области определения $f(x) > f(a)$, а само значение $f(a)$ называется глобальным минимумом.

**Точкой локального минимума** называется такая точка $a$, что для лобого $x$ из некторой окрестности точки $a$ верно, что $f(x) > f(a)$, а само значение $f(a)$ называется локальным минимумом.


Соответственно: если функция $f$ имеет экстремум в точке $a$ и дифференцируема в ней, то $a$ является стационарной точкой функции $f$. 

Так, например, дана кубическая функция $f(x) = x^3$, чья производная показывает нам возрастание функции: $f'(x)=(x^3)'=3x^2 > 0$.  Бесконечно близко около точки $x=0$ скорость изменения функции равна нулю, что подтверждает производная в точке $f(0) = 3 \cdot 0^2 = 0$

Гипербола $f(x) = \frac{1}{x}, f'(x)=-\frac{1}{x^2}< 0$ для любого ненулевого (в нуле функция не определена) аргумента, будет убывать на всей области определения. 

Экспоненциальная функция $f(x) = e^x$ будет возрастать (для любого значения икс справедливо $f'(x)=e^x>0$. Натуральный логарифм тоже: $f(x)=\ln x, f'(x)=\frac{1}{x}$




**Выпуклость и вогнутость графика функции, точки перегиба**

Назовём хордой отрезок, соединяющий две различные точки графика. 

График функции является **выпуклым** на некотором интервале, если он расположен не ниже любой хорды данного интервала.

График функции являются **вогнутым** на интервале, если он расположен не выше любой хорды этого интервала.

Точка графика, в которой он меняет выпуклость на вогнутость или вогнутость на выпуклость, называется **точкой перегиба**. 



```julia
plot(x -> x^3 - 15x^2 - 40x, xlims = (-10, 20))
plot!(x -> -700 + 70x)
```

**Асимптоты к графику функции**

**Асимптота** - прямая, обладающая тем свойством, что расстояние от точки кривой до этой прямой стремится к нулю при удалении точки вдоль ветви в бесконечность. То есть это прямая, к которой неограниченно близко приближается график функции при удалении его переменной точки в бесконечность, но никогда не пересекает ее.

1) Вертикальные асимптоты, которые задаются уравнением вида $x=\alpha$, где «альфа» – действительное число. Ассимптота $x=0$ определяет саму ось ординат, можно вспомнить гиперболу $y = \frac{1}{x}$.

2) Наклонные асимптоты традиционно записываются уравнением прямой с угловым коэффициентом $y=kx+b$. Иногда отдельной группой выделяют частный случай – горизонтальные асимптоты $y=b$. Например, та же гипербола с асимптотой $y=0$.

График функции может не иметь ассимптот, а может иметь их бесконечно много. Найти ассимптоты - это найти их уравнения и начертить соответсвенные на графике.

**Вертикальная ассимптота:** Вертикальная асимптота графика, как правило, находится в точке бесконечного разрыва функции. Всё просто: если в точке $x=\alpha$ функция $y=f(x)$ терпит бесконечный разрыв, то прямая, заданная уравнением $x=\alpha$ является вертикальной асимптотой графика.

Таким образом, чтобы установить наличие вертикальной асимптоты  в точке  достаточно показать, что хотя бы один из односторонних пределов $\lim\limits_{x \to \alpha - 0}f(x), \lim\limits_{x \to \alpha + 0}f(x)$ бесконечен. 

Чаще всего это точка, где знаменатель функции равен нулю. Из вышесказанного также следует очевидный факт: <u>если функция непрерывна на $R$, то  вертикальные асимптоты отсутствуют.<u>

Но в ряде случаев существует только один односторонний предел, и, если он бесконечен, то встречается вертикальная ассимптота (например $f(x)=\ln x$, где ассимптотой выступает ось ординат (y)).


```julia
plot(x->log(x), label="y = ln(x)")
```

**Наклонная ассимптота:** 
Если существуют два конечных предела $\lim\limits_{x \to \pm\infty}\frac{f(x)}{x}=k, \lim\limits_{x \to \pm\infty}(f(x)-kx)=b$ ,  то прямая  является наклонной асимптотой графика функции $y=f(x)$ при $x \to \pm \infty$. Если хотя бы один из перечисленных пределов бесконечен, то наклонная асимптота отсутствует.

Примечание: формулы остаются справедливыми, если $x$ стремится только к $+\infty$ или только к $-\infty$.

Покажем, что у параболы $y=x^2$ нет наклонных ассимптот. 

$k = \lim\limits_{x \to \pm\infty}\frac{f(x)}{x} = \lim\limits_{x \to \pm\infty}\frac{x^2}{x} = \lim\limits_{x \to \pm\infty}x = \pm\infty$

Предел бесконечен, значит, наклонная асимптота отсутствует. Нам также не придется искать коэффициент b, поскольку ответ уже получен.

**Четность** 

Определения вводятся для любой симметричной относительно нуля области определения $X ⊂ R$, например, отрезка или интервала.

Функция $f : X → R$  называется чётной, если справедливо равенство:

* $f ( − x ) = f ( x ) , ∀ x ∈ X $.

Функция называется нечётной, если справедливо равенство:

* $f ( − x ) = − f ( x ) , ∀ x ∈ X$. 


Функции, не принадлежащие ни одной из категорий выше, называются ни чётными, ни нечётными (или функциями общего вида).

Функции, принимающие нулевое значение на всей своей области определения, причём эта область определения симметрична относительно нуля, являются одновременно чётными и нечётными; например, функции $f(x) = 0$ и $f(x) = 0/х$. Любая функция, являющаяся одновременно чётной и нечётной, тождественно равна нулю на всей своей области определения. 



**Периодичность**

Функция $f : M → N$  (где $N$ — произвольное множество её значений) называется периодической с периодом $T ≠ 0$, если справедливо:

$f ( x + T ) = f ( x ) , ∀ x ∈ M$ 

Если это равенство не выполнено ни для какого $T ∈ M$ , $T ≠ 0$ , то функция $f$ называется апериодической. 

**Монотонная функция**

Пусть дана функция $f : M ⊂ R → R$.  Тогда

* функция $f$ называется **возрастающей (неубывающей)** на $M$, если $∀ x , y ∈ M , x > y ⇒ f ( x ) ≥ f ( y ) $.


* функция $f$ называется **строго возрастающей** на $M$, если $∀ x , y ∈ M , x > y ⇒ f ( x ) > f ( y )$.


* функция $f$ называется **убывающей (невозврастающей)** на $M$, если $∀ x , y ∈ M , x > y ⇒ f ( x ) ≤ f ( y )$.


* функция $f$ называется **строго убывающей** на $M$, если $∀ x , y ∈ M , x > y ⇒ f ( x ) < f ( y )$.


(Строго) возрастающая или убывающая функция называется (строго) монотонной
