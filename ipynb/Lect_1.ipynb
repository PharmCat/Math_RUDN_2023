### Модель Лотки-Вольтерры



где x — количество жертв (🐇), y — количество хищников (🦊),  t — время, α , β , γ , δ — коэффициенты, отражающие взаимодействия между видами. 

```julia
using DifferentialEquations, Plots

function lv!(du, u, p, t)
    α, β, γ, δ = p
    
    🐇, 🦊    = u
    
    du[1] =  (α - β * 🦊) * 🐇
    du[2] = (-γ + δ * 🐇) * 🦊

end
# Начальные условия
u0 = [1.0;2.0]
# Параметры
p  = [0.2, 0.2, 0.9, 0.2]
# Отрезок времени
tspan = (0.0, 100.0)
# Задача
prob = ODEProblem(lv!, u0, tspan, p)
# Решение
sol = solve(prob)
# График
plot(sol; label = ["жертвы" "хищники"])
```

### Множества

В широком смысле, множество – это совокупность объектов (элементов), которые понимаются как единое целое (по тем или иным признакам, критериям или обстоятельствам). 

Как задается множество:

Путем перечисления:

$A = \{1, 2, 3, 4, 5\}$

С помощью характеристического свойства:

$A = \{ n ∈ N | n < 6\}$ 

**Мощность множества** - размер множества, т.е. количество его элементов, обозначается $|A|$ или $ \#A $. Множество может быть конечным или бесконечным.

**Пустое множество:**

$∅ = \{\}$

Мощность пустого множества $|∅| = 0$.

**Принадлежность элемента множеству:**

$1 ∈ A$

$6 ∉ A$

Если задать множество $B = \{1, 2, 3\}$, то можно сказать что $B$ является подмножеством $A$ и записать это так:

$B ⊆ A$ (или $B ⊂ A$)

**Универсальное множество**

 $\mathbb{U}$ - универсальное множество, содержащее все объекты и все множества, такое что $¬\mathbb{U} = ∅$.

```julia
# создадим пустой массив (пустое множество)
∅ = []

A  = [1, 2, 3, 4, 5, 6] 

B  = [1, 2, 3] 

println("Пустое множество входит в любое другое: ", ∅ ⊆ A)

println("Элемент 1 входит в множество A: ", 1 ∈ A)

println("Элемент 6 не входит в множество B: ", 6 ∈ B)

println("Множество B входит в множество A: ", B ⊆ A)
```

Также можно использовать обозначения $⊈$ - не входит. $B⊊A$ - означает, что B является подмножеством A, но не равно ему, к примеру: 

```julia
C = [1,2]

C ⊊ B
```

**Произведением** двух множеств A и B называют множество A × B , состоящее из всех упорядоченных пар, первые элементы которых выбираются из A, вторые - из B (т.е. A × B = {(a, b) : a ∈ A, b ∈ B}).

**Соответствием** ϕ между множествами A и B называется произвольное подмножество их произведения A × B (т.е. ϕ ⊆ A × B )

**Областью определения** соответствия ϕ называется множество *Dom* ϕ = {a ∈ A: ∃ b ∈ B, что (a, b) ∈ ϕ} (т.е. это все элементы из A , которым соответствует хотя бы один элемент из B).

**Множеством значений** соответствия ϕ называют множество *Im* ϕ = {b ∈ B: ∃ a ∈ A, что (a, b) ∈ ϕ} (т.е. это все элементы из B, которые соответствуют хотя бы одному элементу из A ).


**Соответствие ϕ называется:**


1) всюду определенным, если *Dom* ϕ = A;


2) сюръективным, если *Im* ϕ = B ;


3) однозначным, если каждому a ∈ *Dom* ϕ соответствует единственный элемент b из B, т.е. из (a, b) ∈ ϕ и (a, $b_1$) ∈ ϕ ⇒ b = $b_1$;


4) инъективным, если разным элементам из *Dom* ϕ соответствуют разные элементы из B, т.е. из (a, b) ∈ ϕ и ($a_1$, b) ∈ ϕ ⇒ a = $a_1$ .


**Отображение множества  во множество** – это правило, по которому каждому элементу множества A ставится в соответствие элемент (или элементы) множества B (всюду определенное и однозначное соответствие (выполняются свойства 1 и 3)). В том случае если в соответствие ставится единственный элемент, то данное правило называется однозначно определённой функцией или просто функцией.

Обозначается: $f: A → B$ или $g: R^2 → R$


**Биекцией** называют всюду определенное, сюръективное, однозначное и инъективное соответствие (выполняются все свойства 1-4).

### Операции над множествами

* Пересечение ($∩$) множеств (логическая связь И);

* Объединение ($∪$) множеств (логическая связь ИЛИ);

* Разность ( ∖ ) - $A$ ∖ $B$ означает элементы принадлежат A, но не принадлежат B; Функция `setdiff`;

* Симметричная разность($⊖$ или иногда $△$) Функция `symdiff` (операция, результатом которой является новое множество, включающее все элементы исходных множеств, не принадлежащие одновременно обоим исходным множествам).

```julia
# Пересечение
A ∩ B
```

```julia
# Объединение
D = [4,5,6,7,8,9]
B ∪ D
```

```julia
# Разность
setdiff(A, D)
```

```julia
# Симметричная разность
symdiff(A, D)
```

**Дополнением** множества A называется разность между универсальным множеством $\mathbb{U}$ и множеством A  ( обозначается $\overline{A}$)

**Некоторые равенства** 

$A ∪ A = A$
    
$A ∩ A = A$
    
$A ∪ B = B ∪ A$
    
$A ∩ B = B ∩ A$
    
$(A ∪ B) ∪ C = A ∪ (B ∪ C)$
    
$(A ∩ B) ∩ C = A ∩ (B ∩ C)$
    
$(A ∪ B) ∩ C = (A ∩ C) ∪ (B ∩ C)$
    
$(A ∩ B) ∪ C = (A ∪ C) ∩ (B ∪ C)$
    
$A ∪ ∅ = A$
    
$A ∩ ∅ = ∅$

$A ∪ \mathbb{U} = \mathbb{U}$

$A ∩ \mathbb{U} = A$

$\overline{A ∪ B} = \overline{A} ∩ \overline{B}$

$\overline{A ∩ B} = \overline{A} ∪ \overline{B}$

$\overline{\overline{A}} = A$
    
$(A ⊂ B) · (B ⊂ C) ⇒ A ⊂ C$

### Логика

Логическим высказыванием считается любое высказывание, про которое можно сказать истинно оно или ложно.

Существуют следующие бинарные операции на атомарных $X$ и $Y$:

* Коньюнкция - смысл "И", обозначение: $X ∧ Y$ ,  $X · Y$ ,  $X  \&\&  Y$
* Дизюнкция - смысл "ИЛИ", обозначение: $X ∨ Y$  ,  $X + Y$  ,  $X  ||  Y $
* Отрицание - смысл "НЕ", обозначение: $ ¬ X$  ,  $\overline{X}$ , $ !X$ 
* Импликация - смысл "СЛЕДУЕТ", обозначение: $X → Y$
* Эквивалентность - смысл "ЭКВИВАЛЕНТНО", обозначение: $X ⬌ Y$
* Исключающее или - смылс "ИЛИ X ИЛИ Y", обозначение: $X ⊕ Y$ , $X ⊻ Y$

Некоторые законы логики:

* Двойное отрицание $ ¬(¬ X) = X$
* Законы Де Моргана $ \overline{X · Y} = \overline{X} + \overline{Y}$;  $ \overline{X + Y} = \overline{X} · \overline{Y}$

Законы Де Моргана:

* Отрицание конъюнкции есть дизъюнкция отрицаний.
* Отрицание дизъюнкции есть конъюнкция отрицаний.

```julia
#Некоторые примеры логических операций


X = true
Y = false

if X println("X - true") end

if X && Y println("X ∧ Y - true") else println("X ∧ Y - false") end

if X || Y println("X ∨ Y - true") else println("X ∨ Y - false") end

if !X println("¬ X - true") else println("¬ X - false") end

if X ⊻ Y println("X ⊕ Y - true") else println("X ⊕ Y - false") end

if !(X && Y) == !X || !Y println("¬(X ∧ Y) = ¬X ∨ ¬Y") end


```
