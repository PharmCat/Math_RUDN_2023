## Производная функции нескольких переменных

**Определение**

Существует обобщение понятия функции нескольких переменных, когда в качестве значений
отображения используются не числа, а векторы (или точки n-мерного пространства). 

В общем случае мы называем функцией многих переменных (функцией нескольких
переменных) отображение вида $f : A → R^m$, где $A ⊂ R^n, n > 1$. Если $m$ = 1, т.е. значени-
ем отображения является действительное число (скалярная величина), отображение называют
**скалярной функцией нескольких переменных**. Если же $m > 1$, то указанное отображение
называют **векторной функцией нескольких переменных** (или векторной функцией
векторного аргумента).


Множество $D(f ) = A ⊂ R^n$, на котором определена функция $f : A ⊂ R^n → R^m$, 
называют областью определения (существования) функции $f$ , а множество 
$R(f ) = {y ∈ R^m: y = f (x), x ∈ D(f )}$ — областью значений (изменения) функции $f$.

**Частные производные** 

В математическом анализе частная производная (первая производная) — одно из обобщений понятия производной на случай функции нескольких переменных. Частная производная — это предел отношения приращения функции по выбранной переменной к приращению этой переменной, при стремлении этого приращения к нулю. 
Иными словами, частная производая - это первая производная, взятая только по одной переменной. Остальные переменные в дифференцируются как константы. Для частных производных справедливы все правила дифференцирования и таблица производных элементарных функций.

Производная функции нескольких переменных будет обозначается как ${\displaystyle {\frac {\partial f(x_{1}~...~x_{k}~...~x_{n})}{\partial x_{k}}}}$, что есть её дифференцирование по одному из аргументов:

$${\frac {\partial f(x_{1}~...~x_{k}~...~x_{n})}{\partial x_{k}}}=\lim _{\Delta x\to 0}{\frac {f(x_{1}~...~x_{k}+\Delta x_{k}~...~x_{n})-f(x_{1}~...~x_{k}~...~x_{n})}{\Delta x_{k}}}$$

где $x_k$ — определенный аргумент; а символ $\partial$ является видоизмененной записью $d_{x}$. 

Таким образом, частная производная функции $f$ по переменной $x$ обычно обозначается ${\displaystyle {\tfrac {\partial f}{\partial x}}}$, $f_{x}$ или ${\displaystyle D_{x}f}$. 


**Пример 1.** 

Найти производную первого и второго порядка для $z = 2x^2y^3 + 3x^4 + 5y -7$

Частных производных будет две: 

* $z^{'}_x$ или $\frac{\partial z}{\partial x}$ - частная производная по $x$

* $z^{'}_y$ или $\frac{\partial z}{\partial y}$ - частная производная по $y$


Найдем частную производную по $x$. Тогда переменная $y$ будет считаться константой (постоянным числом).
Как мы знаем, производная от константы - 0.

$z^{'}_x = (2x^2y^3 + 3x^4 + 5y -7)^{'} = 2y^3(x^2)^{'}_x + 3(x^4)^{'}_x + (5y)^{'}_x -(7)^{'}_x$ 

Теперь, применяя все правила дифференцирования и считая $y$ константой, решаем:

$z^{'}_x = 2y^3 \dot 2x + 3 * 4x^3 + 0 - 0 = 4xy^3 + 12x^3$ 

Обратите внимание, что так как $y$ считается константой, а любую константу можно вынести за знак производной, то $y^3$  мы выносим за скобки. То есть в данной ситуации $y^3$ можно назвать обычным числом. Теперь посмотрим на третье слагаемое $5y$: здесь, наоборот, выносить нечего. Так как $y$ константа, то  – $5y$ тоже константа, и будет эквивалентна последнему слагаемому - цифре 7. 

Теперь по аналогии найдем частную производную по игрек. В данном случае уже $x$ становится константой. 

$z^{'}_y = (2x^2y^3 + 3x^4 + 5y -7)^{'} = 2x^2(y^3)^{'}_y + 3(x^4)^{'}_y + (5y)^{'}_y -(7)^{'}_y 
= 2x^2 \dot 3y^2 + 0 + 5 - 0 = 6x^2y^2 + 5 $

**В чем смысл частных производных?** 

По своей сути частные производные 1-го порядка напоминают «обычную» производную: $z^{'}_x$, $z^{'}_y$ – это функции, которые характеризуют скорость изменения функции $z = f(x, y)$ в направлении осей $OX$ и $OY$ соответственно. Так, например, функция $z^{'}_x = 4xy^3 + 12x^3 $ характеризует крутизну «подъёмов» и «склонов» поверхности $z = 2x^2y^3 + 3x^4 + 5y -7$ в направлении оси абсцисс, а функция $z^{'}_y = 6x^2y^2 + 5$ сообщает нам о «рельефе» этой же поверхности в направлении оси ординат.


**Полный дифференциал функции двух переменных**

Частные производные необходимы нам для поиска полного дифференциала. Дифференциал - линейная часть приращения функции и имеет вид: $dZ = z^{'}_xdx + z^{'}_ydy$ или $dZ = \frac{\partial Z}{\partial x}dx + \frac{\partial Z}{\partial y}dy$. В данном случае $dx$ и $dy$ - просто символы дифференциала, и для нахождения полного дифференциала нам необходимо сложить частные производные функции. 

Таким образом, продолжая пример выше, мы получим полный дифференциал функции двух переменных: 

$dz = z^{'}_xdx + z^{'}_ydy = (4xy^3 + 12x^3)dx + (6x^2y^2 + 5)dy$

```julia
using ForwardDiff, Plots

f(x) =  x[1]^2 + 1.5x[2]^2 + 2x[2] + log(10x[1]^2 + 0.1)*x[2]
fp(x, y) = f([x, y])
x = y = -5:0.01:5
z = fp.(x', y)
surface(x, y, z)
```

```julia
#используя ForwardDiff найти производные функции f 

g = ForwardDiff.gradient(f, [2,4])
```

### Градиент

**Производная по направлению**

Если производная по $y^{'}$ - это скорость изменения функции $y = f(x)$ в направлении оси $OX$, то есть грубо говоря "по горизонтали", то при данной нам функции $z = f(x,y)$ мы можем охарактеризовать ее еще по оси $OY$. 

Чтобы узнать скорость изменения функции $z = f(x, y)$ в каком либо направлении, мы будем пользоваться производной по направлению $\frac{\partial z}{\partial l}$. В качестве направления выступает вектор $l$.

Представим себе точку $M_0(x_0;y_0)$ из области определения функции. Из этой точки мы можем двигаться в сторону $OX$ и $OY$, и направление нашего движения будет обозначаться лучом $l$, исходящим из точки $M_0$. Сам луч можно опрелелить с помощью угла между ним и любой осью, либо с помощью вектора. 

**Определение:**

Производной $\frac{\partial f}{\partial l}$ функции $f$ в точке $M_0$ по направлению вектора $l$ называется предел отношения $\frac{f(M) - f(M_0)}{|l|}$, при $|l| \to 0$.


**Пример**
Eсли в точке $M_0(x_0;y_0)$ существует такая производная по направлению луча, то мы можем рассчитать ее по формуле: 

$$\frac{\partial z}{\partial l} = z^{'}_x(x_0;y_0)  \cos\alpha + z^{'}_y(x_0;y_0) \cos\beta$$, 

где $z^{'}_x(x_0;y_0)$ и $z^{'}_y(x_0;y_0)$ - частные производные первого порядка в точке $M_0$; $\cos\alpha$, $\cos\beta$ - направляющие косинусы (координаты вектора единичной длины), однозначно определяющие данное направление. 

**Общая формула вычисления производной по направлению:**

$$\frac{\partial f}{\partial l}(M)= \frac{f_{x_1}'(M)l_1 + f_{x_2}'(M)l_2 + ... + f_{x_n}'(M)l_n}{|l|}$$

Так как $\frac{\partial z}{\partial l}$ - это число, то результат нашего решения можно интерпретировать: 
- если $\frac{\partial z}{\partial l} > 0$, то функция в точке $M_0$ по данному направлению возрастает; 
- если $\frac{\partial z}{\partial l} < 0$, то функция в точке $M_0$ по данному направлению убывает; 
- если $\frac{\partial z}{\partial l} = 0$, то функция в точке $M_0$ по данному направлению постоянна (параллельна плоскости); 



**Пример 2.** 

Найти производную функции $z = x - y$ в точке $M_0(0;0)$ по направлению вектора $\vec{a}(2;2)$.


Вычислим скорость изменения функции по направлению, исходящего из точки $M_0(0;0)$ луча $l$ по формуле: 
$\frac{\partial z}{\partial l} = z^{'}_x(x_0;y_0) \cos\alpha + z^{'}_y(x_0;y_0) \cos\beta$.

Найдем частные производные:
\begin{equation}
    z^{'}_x = (x)^{'}_x - (y)^{'}_x = 1 - 0 = 1
\end{equation}
\begin{equation}
z^{'}_y(x_0;y_0) = (x)^{'}_y - (y)^{'}_y = 0 - 1 = -1
\end{equation}

Подставляем в формулу: 
\begin{equation}
    \frac{\partial z}{\partial l} = z^{'}_x(M_0)  \cos\alpha + z^{'}_y(M_0)  \cos\beta =1 \dot \cos \alpha + (-1)\dot \cos \beta = \cos\alpha-\cos\beta
\end{equation}

Теперь требуется найти вектор $\vec{a_0}(\cos\alpha;\cos\beta)$ единичной длины, сонаправленный с вектором $\vec{a}(2;2)$ по формуле нахождения скалярного произведения векторов.

\begin{equation}
\vec{a_0}=\frac{\vec{a}}{|\vec{a}|} = \frac{2\vec{i}+2\vec{j}}{\sqrt{2^2 + 2^2}}= \frac{2\vec{i}+2\vec{j}}{2\sqrt{2}}= \frac{1}{\sqrt{2}\vec{i}}+ \frac{1}{\sqrt{2}\vec{j}}
\end{equation}

Проверить можно, убедившись, что квадратный корень из суммы квадратов полученных при $\vec{i}$ и $\vec{j}$ точек равен единице. 

Теперь подставляем в формулу: 

\begin{equation}
    \frac{\partial z}{\partial l} = z^{'}_x(M_0)  \cos\alpha + z^{'}_y(M_0)  \cos\beta = 1 *\frac{1}{\sqrt{2}} + (-1) * \frac{1}{\sqrt{2}} = 0 
\end{equation}

Скорость изменения функции постоянна вне зависимости от того, от какой точки в плоскости $z = x - y$ по направлению вектора $\vec{a}(2;2)$  мы будем двигаться. 

**Градиент**

Градиентом функции нескольких переменных $f$ в точке $M$ называется вектор:

$$grad f(M) = \nabla f(M) = (f_{x_1}'(M), \ f_{x_2}'(M), \ ... \ , \ f_{x_n}'(M))$$ 


Вектор $\nabla f(M)$ указывает на направление наиболее быстрого роста функции в точке $M$, а его длина равна скорости этого роста.

Производная функции в направлении, перпендикулярному вектору $\nabla f(M)$, равна нулю.

В каждой точке градиент перпендикулярен линии (поверхности) уровня, проходящей через эту точку.

**Стационарная (критическая) точка**

Точка, в которой все частные производные первого порядка равны нулю  ($\nabla f(M) = 0$), называется стационарной.



**Экстремумы функции нескольких перменных**

* Точка $M_0$ называется точкой локального **минимума** функции $f$, если существует такая окрестность $M_0$, что в любой точке $M$ этой окрестности выполняется неравенство $f(M_0) ≤ f(M)$.

* Точка $M_0$ называется точкой локального **максимума** функции $f$, если существует такая окрестность $M_0$, что в любой точке $M$ этой окрестности выполняется неравенство $f(M_0) ≥ f(M)$.

Для решения задач на определение экстремума необходимо разобрать понятие о второй производной.

*Окрестность точки — множество, содержащее данную точку и близкие к ней.

**Вторая производная**

Вторая производная простым языком, это производная первой производной. Так как для функции многих переменных количество частных производных равно количеству аргументов функции, то и вторые производные это также частные производные получившихся функций многих переменных.

К примеру:

$z = 2x^2y^3 + 3x^4 + 5y -7$

Первые производные:

* $z^{'}_x = \frac{\partial z}{\partial x} = 4xy^3 + 12x^3$

* $z^{'}_y = \frac{\partial z}{\partial y} = 6x^2y^2 + 5$ 


Соответственно вторые производные:

* $\frac{\partial z}{\partial x} = 4xy^3 + 12x^3$
    * $\frac{\partial^2 z}{\partial x^2} = 4y^3 + 36x^2$
    * $\frac{\partial^2 z}{\partial x \partial y} = 12xy^2$
    

* $\frac{\partial z}{\partial y} = 6x^2y^2 + 5$ 
    * $\frac{\partial^2 z}{\partial y^2} = 12x^2y$
    * $\frac{\partial^2 z}{\partial y \partial x} = 12xy^2$


Производные $\frac{\partial z}{\partial y \partial x}$  и $\frac{\partial z}{\partial x \partial y}$  называются смешанные производные и равны друг другу.

**Матрица Гессе**


Матрица этой квадратичной формы образована вторыми частными производными функции. Если все производные существуют, то

$$ \nabla^2 f = H(f)={\begin{bmatrix}{\frac {\partial ^{2}f}{\partial x_{1}^{2}}}&{\frac {\partial ^{2}f}{\partial x_{1}\,\partial x_{2}}}&\cdots &{\frac {\partial ^{2}f}{\partial x_{1}\,\partial x_{n}}}\\\\{\frac {\partial ^{2}f}{\partial x_{2}\,\partial x_{1}}}&{\frac {\partial ^{2}f}{\partial x_{2}^{2}}}&\cdots &{\frac {\partial ^{2}f}{\partial x_{2}\,\partial x_{n}}}\\\\\vdots &\vdots &\ddots &\vdots \\\\{\frac {\partial ^{2}f}{\partial x_{n}\,\partial x_{1}}}&{\frac {\partial ^{2}f}{\partial x_{n}\,\partial x_{2}}}&\cdots &{\frac {\partial ^{2}f}{\partial x_{n}^{2}}}\end{bmatrix}}$$

Определитель этой матрицы называется определителем Гессе, или гессианом.

```julia
#используя ForwardDiff найти матрицу Гессе f 

h = ForwardDiff.hessian(f, [2,4])
```

```julia
using LinearAlgebra
det(h)
```

**Достаточное условие экстремума**

Пусть $|\nabla^2 f(M)|$  - определитель  матрицы вторых производных (гессиан) в точке $M$.

Если $|\nabla^2 f(M)| > 0$, то функция $f$ имеет экстремум в этой точке. 

**При этом для функции двух переменных:**

Для функции двух переменных достаточно что бы первый диагональный элемент $\nabla^2 f(M)$ был > 0, тогда в этой точке достигается минимум. Если первый диагональный элемент $\nabla^2 f(M)$ < 0 - максимум. 

**Для функции многих переменных:**

* если матрица $\nabla^2 f(M)$ по критерию Сильвестра положительно определенная, то в точке  функция  имеет локальный минимум;

* если матрица отрицательно определенная, то в точке  функция  имеет локальный максимум;

* если матрица неопределенная, то в точке  функция  не имеет локального экстремума (это – так называемая седловая точка).


Если $|\nabla^2 f(M)| < 0$, то функция $f$ не имеет экстремума в этой точке (седловая точка). 

*Симметричная матрица отрицательно(положительно) определена если все собсвенные числа отрицательные(положительные).*

Пусть матрица имеет вид:

   $A={\begin{matrix}a_{11}&a_{12}&\ldots &a_{1n}\\a_{21}&a_{22}&\ldots &a_{2n}\\\ldots &\ldots &\ldots &\ldots \\a_{n1}&a_{n2}&\ldots &a_{nn}\end{matrix}}$

Тогда матрица положительно определена тогда и только тогда, когда все её угловые миноры $\Delta _{i}$ размеров $i × i$, где $i = 1 ... n$, положительны; а отрицательно определена тогда и только тогда, когда знаки $\Delta _{i}$ чередуются, причём $ \Delta _{1}<0$. Угловыми минорами матрицы $A$ называются определители вида

$\Delta _{1}=a_{11},\ \Delta _{2}={\begin{vmatrix}a_{11}&a_{12}\\a_{21}&a_{22}\end{vmatrix}},\ldots ,\Delta _{i}={\begin{vmatrix}a_{11}&a_{12}&\ldots &a_{1i}\\a_{21}&a_{22}&\ldots &a_{2i}\\\ldots &\ldots &\ldots &\ldots \\a_{i1}&a_{i2}&\ldots &a_{ii}\end{vmatrix}},\ldots $

**Матрица Якоби**

Матрица Якоби векторной функции  $\mathbf {u} :\mathbb {R} ^{n}\to \mathbb {R} ^{n},\mathbf {u} =(u_{1},\ldots ,u_{n}),u_{i}=u_{i}(x_{1},\ldots ,x_{n}),i=1,\ldots ,n$  имеющей в некоторой точке $x$ все частные производные первого порядка, определяется как:

$$ J = {\begin{pmatrix}{\partial u_{1} \over \partial x_{1}}(x)&{\partial u_{1} \over \partial x_{2}}(x)&\cdots &{\partial u_{1} \over \partial x_{n}}(x)\\{\partial u_{2} \over \partial x_{1}}(x)&{\partial u_{2} \over \partial x_{2}}(x)&\cdots &{\partial u_{2} \over \partial x_{n}}(x)\\\cdots &\cdots &\cdots &\cdots \\{\partial u_{n} \over \partial x_{1}}(x)&{\partial u_{n} \over \partial x_{2}}(x)&\cdots &{\partial u_{n} \over \partial x_{n}}(x)\end{pmatrix}}.$$

Также можно говорить об определителе Якоби или якобиане системы функций $u_{1},\ldots ,u_{n}.$


**Решение задач оптимизации методом градиентного спуска**

Метод градиентного спуска:

$$X_{k+1} = X_k - t_k \nabla f(X_k)$$


$k$ выбирается как (метод наискорейшего спуска):

$$k = argmin_k f(X_k - t_k \nabla f(X_k))$$

Пример:

$$f(x,y) = 3x^2 + y^2 - 2xy \to min$$

$x_0 = 4; y_0 = 2$


$$\frac{\partial f}{\partial x} = 6x - 2$$

$$\frac{\partial f}{\partial y} = 2y - 2$$



$\frac{\partial f}{\partial x} = 6 * 4 - 2 = 22$

$\frac{\partial f}{\partial y} = 2 * 2 - 2 = 2$

$$\nabla f(x_0, y_0) = [22, 2]$$

$$k \nabla f(x_0, y_0) = [22k, 2k]$$

$$[4, 2] - k \nabla f(x_0, y_0) = [4 - 22k, 2 - 2k]$$

$ k = argmin_k f(X_k - t_k \nabla f(X_k)) = f(4 - 22k, 2 - 2k)= 3(4 - 22k)^2 + (2 - 2k)^2 - 2(4 - 22k)(2 - 2k)$

$f'(X_k - t_k \nabla f(X_k)) = 2736·k-432 = 0$

$k = 432 / 2736 ≈ 0.15789$
