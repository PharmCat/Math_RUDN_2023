### Обыкновенное дифференциальное уравнение

Обыкновенное дифференциальное уравнение $n$-го порядка представляет собой уравнение вида:


$$f(x, y,y',y'',\ldots ,y^{(n)}) = 0$$


Или в следующей записи:


$$y^{(n)} = f(x, y,y',y'',\ldots ,y^{(n-1)}),$$


где функция $f$, называется разрешенной относительно старшей производной.

**Порядком** дифференциального уравнения называется наивысший порядок производной (или дифференциала) неизвестной функции, входящей в уравнение.

### Дифференциальные уравнения  допускающие понижение порядка

Существуюет несколько видов дифференциальных уравнений  допускающих понижение порядка.

ДУ  допускающие понижение порядка могут быть записаны в виде:

1) Вида:

$$y^{(n)} = f(x)$$

где $f(x)$ – заданная функция, решаются непосредственным последовательным интегрированием.

2) Вида:

$$f(x, y',y'',\ldots ,y^{(n)}) = 0,$$

не содержат явно искомую функцию $y$, допускают понижение порядка подстановкой $y' = p(x)$


3) Вида:

$$f(y, y',y'',\ldots ,y^{(n)}) = 0,$$

не содержат явно переменную $x$, допускают понижение порядка подстановкой $y' = p(y)$, тогда:

$$y'' = (y')' = (p(y))' = (p(y(x)))'$$

Помня, что $(u(v))' = u'(v)· v'$ и $y' = p(y)$

$$(p(y))' = p'(y)· y' = p'(y)· p = p' · p $$

В общем виде:

$$y^{(n)} = g(p, p',p'',\ldots ,p^{(n-1)}),$$

где:

$$p^{(i)} = \frac{\partial^i p}{\partial y^i}, i = 1,2,\ldots,n-1$$

К примеру:

$$y''' = (y'')' = (p' · p)' = p''· p^2 + (p')^2· p$$

4) Вида:

$$\frac{d}{dx}f(y, y',y'',\ldots ,y^{(n-1)}) = 0,$$

в которых левая часть может быть представлена как полная производная от некоторой функции $f(y, y',y'',\ldots ,y^{(n-1)})$

### ДУ допускающие понижение порядка вида 1

$$y''' = sin(x)$$

Тогда:

$$(y'')' = sin(x)$$

$$\frac{d (y'')}{dx} = sin(x)$$

$$d(y'') = sin(x)dx$$

$$\int d(y'') = \int sin(x)dx$$

$$y'' = - cos(x) + C_1$$


$$(y')' = - cos(x) + C_1$$

$$\frac{d(y')}{dx} = - cos(x) + C_1$$

$$d(y') = (-cos(x) + C_1)dx$$

Тогда:

$$y' = -sin(x) + C_1x + C_2$$

$$\frac{dy}{dx} = -sin(x) + C_1x + C_2$$

$$dy = (-sin(x) + C_1x + C_2)dx$$

общее решение данного уравнения будет иметь вид

$$y = cos(x) + \frac{x^2}{2}C_1 + C_2x + C_3$$

### ДУ допускающие понижение порядка вида 2

$$y'' = 5 · y' · \frac{1}{x}$$

Данное уравнение не содержит искомой функции $y$, поэтому для его решения проведем замену
$y′ = p(x)$ и $y′′ = p′(x)$:


$$p' = 5 · p · \frac{1}{x}$$

$$\frac{dp}{dx} = 5 · p · \frac{1}{x}$$

Интегрируя:

$$\frac{dp}{p} = 5 · \frac{dx}{x}$$

Получаем:

$$ln|p| = 5ln|x| + ln(C_1)$$

$$p(x) = C_1 ⋅ x^5$$



Помня, что $y′ = p(x)$, подставляем $p(x) = C_1 ⋅ x^5$ в это уравнение получаем:


$$y′ = C_1 ⋅ x^5,$$

Далее решаем обычное ДУ первого порядка:


$$\frac{dy}{dx} = C_1 ⋅ x^5$$

$$dy = C_1 ⋅ x^5 ⋅ dx$$

$$y = \frac{C_1 ⋅ x^6}{6} + C_2$$

$$y = \hat{C_1} ⋅ x^6 + C_2$$

### ДУ допускающие понижение порядка вида 3

$$(y')^2+ 2⋅y⋅y''  = 0,$$

Данное уравнение не содержит явно независимой переменной $x$ , а потому относится к рассматриваемому типу.

Выполним замену $y' = p$ и помним, что $$y'' = p' · p $$, тогда :

$$p^2 + 2 ⋅ y ⋅ p' · p = 0$$

Разделим обе части последнего уравнения на $p$:

$$p + 2 ⋅ y ⋅ p' = 0$$

$$p + 2 ⋅ y ⋅ \frac{dp}{dy} = 0$$

$$ 2 ⋅ y ⋅ \frac{dp}{dy}  = - p$$

$$-\frac{1}{y}dy  = 2\frac{1}{p} dp $$

$$2ln|p| = −ln|y| + ln(C_1^2)$$


$$p^2 = \frac{C_1^2}{y}$$


$$p = \frac{C_1}{\sqrt{y}}$$


С учетом того, что $p = y' = \frac{dy}{dx}$ получаем:

$$\frac{dy}{dx} = \frac{C_1}{\sqrt{y}}$$

$$\sqrt{y} dy  = C_1 dx $$

$$\frac{2}{3} ⋅ y^{\frac{3}{2}}  = C_1⋅x + C_2$$

### ДУ допускающие понижение порядка вида 4

Уравнение:

$$y'' y - (y')^2 = 4x^3y^2$$

$$\frac{ y'' y - y' y'}{y^2} = 4x^3$$

Помня: $(\frac{u}{v})' = (\frac{u' · v - u · v'}{v^2})$

$$\frac{ y'' y - y' y'}{y^2} = (\frac{y'}{y})' = 4x^3$$

$$\int (\frac{y'}{y})' = \int 4x^3$$

$$\frac{y'}{y} =  x^4 + C_1$$

#### Линейные однородные дифференциальные уравнения порядка $n$ с постоянными коэффициентами


Уравнения типа:

$$a_0y + a_1y' + a_2y'',\ldots , a_ny^{(n)} = 0,$$

где $a_0,\ldots , a_n$ - числа.

При этом:

Если $y_1, y_2, \ldots , y_n$ – решения уравнения, то и любая
их линейная комбинация $c_1⋅y_1 + c_2⋅y_2 + \ldots + c_n⋅y_n$ также является решением уравнения, где $c_1 , c_2, \ldots, c_n$ – некоторые постоянные.

Любая система из $n$ линейно независимых решений $y_1(x), y_2(x), \ldots , y_n(x)$ линейного однородного уравнения называется **фундаментальной системой решений уравнения**.

**Характеристическим уравнением** для линейного однородного дифференциального уравнения порядка $n$ с постоянными коэффициентами:

$$a_0y + a_1y' + a_2y'',\ldots , a_ny^{(n)} = 0 \ (1)$$

называется алгебраическое уравнение степени n вида:

$$a_0λ^0+ a_1λ^1 + a_2λ^2,\ldots , a_nλ^{(n)} = 0 \  (2) $$

Таким образом, чтобы составить характеристическое уравнение (2), надо в уравнении (1) заменить производные $y^{(n)}$  степенями неизвестной величины $λ$ (показатель степени с основанием $λ$ должен быть равен порядку соответствующей производной неизвестной функции $y$.

1) Если все корни характеристического уравнения  $λ_1, λ_2,\ldots, λ_{n}$ – числа вещественные и не кратные (т. е. среди них нет равных между собой), то функции


$$y_1(x) = e^{λ_1x}, y_2(x) = e^{λ_2x}, \ldots , y_n(x) = e^{λ_nx}$$


образуют фундаментальную систему решений уравнения.

2) Если все корни характеристического уравнения – числа вещественные, но среди них есть кратные (т. е. равные между собой), то каждому корню $λ_i$ кратности $p$ соответствует $p$ линейно независимых функций вида:

$$y_1(x) = e^{λ_ix}, y_2(x) = x⋅e^{λ_ix}, \ldots , y_p(x) = x^{p-1}⋅e^{λ_ix}$$

из фундаментальной системы решений уравнения.

Решить задачу Коши для линейного однородного дифференциального уравнения:


$$y'' + 9y' + 20y = 0, \quad y(0) = 0, \quad y'(0) = −1$$


Составляем характеристическое уравнение вида:

$$λ^2 + 9λ + 20 = 0$$


Корни этого уравнения: $λ_1 = −4, λ_2 = −5$, которые вещественны и не равны между собой (т. е. не кратные). Тогда фундаментальную систему решений (ФСР) образуют функции:

$$y_1(x) = e^{-4x}, y_2(x) = e^{-5x}$$


Тогда общее решение исходного дифференциального уравнения есть линейная комбинация фундаментальных
решений $y_1(x), y_2(x)$, а именно:


$$y(x) = C_1⋅e^{-4x} + C_2⋅e^{-5x}$$

Решить решения задачи Коши сперва найдем производные:

$$y'(x) = -4⋅C_1⋅e^{-4x} -5⋅C_2⋅e^{-5x}$$

Подставим начальные условия  и получим следующую систему уравнений для определения:

\begin{cases} 0 = C_1 + C_2 \\ -1 =  - 4⋅C_1 - 5⋅C_2 \end{cases}


Откуда:


$$C_1 = -1 \quad C_2 = 1$$


Соответственно искомое решение задачи Коши имеет вид:


$$y(x) = -e^{-4x} + e^{-5x}$$

#### Линейные неоднородные дифференциальные уравнения порядка $n$ с постоянными коэффициентами

Линейным неоднородным дифференциальным уравнением порядка $n$ с постоянными коэффициентами называется уравнение вида:


$$a_0y + a_1y' + a_2y'',\ldots , a_ny^{(n)} = f(x)$$


Общее решение $y_{о.н.}(x)$ линейного неоднородного уравнения  равно сумме общего решения $y_{о.о.}(x)$ соответствующего ему линейного однородного уравнения и какого-нибудь частного решения $y_{ч.о.}(x)$ данного неоднородного уравнения, т. е. находится по формуле:


$$y_{о.н.}(x) = y_{о.о.}(x) + y_{ч.о.}(x)$$

#### Метод вариации (изменения) произвольных постоянных (метод Лагранжа)

Сущность этого метода состоит в следующем. 

1) Первоначально решается линейное однородное уравнение соответствующее данному линейному неоднородному уравнению.

2) Получив общее решение $y_{о.о.}(x)$ соответствующего линейного однородного уравнения 

$$y_{о.о.}(x) = C_1⋅y_1 + C_2⋅y_2 + \ldots + C_n⋅y_n$$


(где $C_1 , C_2, \ldots, C_n$ – некоторые постоянные, $y_1 , y_2, \ldots, y_n$  – фундаментальная система решений линейного однородного уравнения) поступают так: полагают, что в решении  величины $C_1 , C_2, \ldots, C_n$  являются не постоянными, а функциями независимой переменной $x$ , и решение линейного неоднородного уравнения ищут в виде:

$$y_{о.н.}(x) = c_1(x)⋅y_1 + c_2(x)⋅y_2 + \ldots + c_n(x)⋅y_n$$


Где функции $c_1(x), c_2(x), \ldots,  c_n(x)$ определяются из системы уравнений:


\begin{cases}  c_1'(x)⋅y_1 + c_2'(x)⋅y_2 + \ldots + c_n'(x)⋅y_n  = 0 \\ 
c_1'(x)⋅y_1' + c_2'(x)⋅y_2' + \ldots + c_n'(x)⋅y_n'  = 0 \\ 
\quad \ldots \quad \quad  \ldots \quad \quad  \ldots \quad \quad  \ldots \quad  \quad \ldots \\
c_1'(x)⋅y_1^{(n-2)} + c_2'(x)⋅y_2^{(n-2)} + \ldots + c_n'(x)⋅y_n^{(n-2)}  = 0 \\ 
c_1'(x)⋅y_1^{(n-1)} + c_2'(x)⋅y_2^{(n-1)} + \ldots + c_n'(x)⋅y_n^{(n-1)}  = f(x) \\ \end{cases}



Относительно функций $c_1(x), c_2(x), \ldots,  c_n(x)$ система  является системой $n$ линейных неоднородных алгебраических
уравнений, причем главный определитель этой системы – **определитель Вронского:**


$$Δ = W = \begin{vmatrix}
y_1 & y_2 & \ldots & y_n  \\
y_1' & y_2' & \ldots & y_n'  \\
\ldots & \ldots & \ldots & \ldots \\
y_1^{n-1} & y_2^{n-1} & \ldots & y_n^{n-1}  
\end{vmatrix} ≠ 0$$

В этом случае система имеет единственное решение.

**Пример**


$$y'' − 2y' − 3y = e^{4x}$$


Соответствующее однородное уравнение:


$$y'' − 2y' − 3y = 0$$


Характеристическое уравнение:


$$λ^2 − 2λ − 3 = 0$$


Уравнение имеет два различных действительных корня: $λ_1 = −1, \quad λ_2 = 3$.


Фундаментальная система решений линейного однородного дифференциального уравнения состоит из функций

\begin{cases} y_1(x) = e^{-x} \\ y_2(x) = e^{3x} \end{cases}

Следовательно, общее решение $y_{о.о.}(x)$ линейного однородного уравнения имеет вид:

$$y_{о.о.}(x) = C_1e^{-x} + C_2e^{3x}$$

Общее решение исходного линейного неоднородного уравнения , согласно методу Лагранжа, будем искать в виде:


$$y_{о.н.}(x) = c_1(x)⋅e^{-x} + c_2(x)⋅e^{3x}$$

```julia
using DifferentialEquations, Plots
using Sundials
```

```julia
# Модель 4, eq 15-18

function pkev(du, u, p, t)
    kₐ, kₑ, k₁₂,k₂₁  = p
    if kₐ < 0 kₐ = 0 end
    
    A, Xc, Xp        = u
    du[1] = - kₐ * A
    du[2] =  kₐ * A - k₁₂ * Xc - kₑ * Xc + k₂₁ * Xp 
    du[3] =  k₁₂ * Xc - k₂₁ * Xp
end
pk_u0 = [10.0, 0.0, 0.0]      # Начальные условия
pk_p  = [0.4, 0.2, 0.1, 0.04] # Параметры
tspan = (0.0, 180.0)          # Отрезок времени
# Задача
prob_pk = ODEProblem(pkev, pk_u0, tspan, pk_p)
# Решение
pkevsol = solve(prob_pk)
pk_sol_f(x) = pkevsol(x)[3]


function pd(du, u, p, t)
    L₀, L₁, ω, κ₁, κ₂    = p
    C = pk_sol_f(t)
    x₁, x₂, x₃, x₄       = u
    ω = x₁ + x₂ + x₃ + x₄
    du[1] = (2L₀*L₁*x₁^2)/(L₁+2L₀*x₁)/ω - κ₂ * C * x₁
    du[2] = κ₂ * C * x₁ - κ₁ * x₂
    du[3] = κ₁ * x₂ - κ₁ * x₃
    du[4] = κ₁ * x₃ - κ₁ * x₄
end
ω₀ = 10.0
pd_u0 = [ω₀, 0.0, 0.0, 0.0]           # Начальные условия
pd_p  = [0.5, 2., 25.0, 0.05, 0.045]  # Параметры
tspan = (0.0, 180.0)                  # Отрезок времени
# Задача
prob_pd = ODEProblem(pd, pd_u0, tspan, pd_p)
# Решение, используется CVode Backward Differentiation Formula (BDF) solver из Sundials.jl
sol = solve(prob_pd, CVODE_BDF())
# График
plot(sol; idxs = [1], label = ["X1"], ylabel="Tumor", xlims = (0, 120), ylims = (0, 20))
plot!(twinx(), pk_sol_f, label = ["PK_P"], ylabel="PK_P", c=:red, ylims = (0, 3))
```

```julia
# Время в которое произойдет изменение состояния системы ДУ
dosetimes = [48.0, 96.0]

# Условие: функция условия - если время находится в перечне dosetimes, то будет выполнено действие
condition(u, t, integrator) = t ∈ dosetimes

# Эффект: функция, которая модифицирует систему ДУ, в данном случае к значениям количества вещества прибавит дозу 10 
affect!(integrator) = integrator.u[1] += 10.0

# Для обработки этих уловий создается специальная структура (подробности: https://docs.sciml.ai/DiffEqDocs/stable/features/callback_functions/)
cb = DiscreteCallback(condition, affect!)

# Решение ДУ
pkevsol = solve(prob_pk, Tsit5(), callback = cb, tstops = dosetimes)

# Функция только для периферического компартмента
pk_sol_f_mult(x) = pkevsol(x)[3]


plot(pk_sol_f, label = ["PK_P"], ylabel="PK_P", c=:red, ylims = (0, 4), xlims = (0, 180))
```

```julia
# Решение
sol = solve(prob_pd, CVODE_BDF())
# График
plot(sol; idxs = [1], label = ["X1"], ylabel="Tumor", xlims = (0, 180), ylims = (0, 20))
plot!(twinx(), pk_sol_f, label = ["PK_P"], ylabel="PK_P", c=:red, ylims = (0, 4))
```

```julia
# Случайный шум

function σ_pkev(du, u, p, t)
    du[1] = u[1]*0.2
    du[2] = 0.0
    du[3] = 0.0
end

prob_sde_pkev = SDEProblem(pkev, σ_pkev, pk_u0, tspan, pk_p)
sde_sol = solve(prob_sde_pkev)
plot(sde_sol, xlims = (0, 24))
```

```julia
using DifferentialEquations.EnsembleAnalysis

ensembleprob = EnsembleProblem(prob_sde_pkev)

ens_sol = solve(ensembleprob, EnsembleThreads(), trajectories = 1000)

summ = EnsembleSummary(ens_sol, 0:0.1:24)

plot(summ, labels = "Middle 95%")

summ = EnsembleSummary(ens_sol, 0:0.1:24; quantiles = [0.25, 0.75])
plot!(summ, labels = "Middle 50%", legend = true)
```

```julia
function pd_sde(du, u, p, t)
    L₀, L₁, ω, κ₁, κ₂    = p[5:9]
    x₁, x₂, x₃, x₄       = u[4:7]
    ω = x₁ + x₂ + x₃ + x₄
    
    kₐ, kₑ, k₁₂,k₂₁  = p[1:4]
    A, Xc, Xp        = u[1:3]
    
    if kₐ < 0 kₐ = 0 end
    
    C = u[3]
    
    if u[4] < 0.01 u[4] = 0.0 end
    
    du[1] = - kₐ * A
    du[2] =  kₐ * A - k₁₂ * Xc - kₑ * Xc + k₂₁ * Xp 
    du[3] =  k₁₂ * Xc - k₂₁ * Xp
    
    du[4] = (2L₀*L₁*x₁^2)/(L₁+2L₀*x₁)/ω - κ₂ * C * x₁
    du[5] = κ₂ * C * x₁ - κ₁ * x₂
    du[6] = κ₁ * x₂ - κ₁ * x₃
    du[7] = κ₁ * x₃ - κ₁ * x₄
end

ω₀ = 10.0
pkpd_u0 = [10.0, 0.0, 0.0, ω₀, 0.0, 0.0, 0.0]                # Начальные условия
pkpd_p  = [0.4, 0.2, 0.1, 0.04, 0.5, 2., 25.0, 0.05, 0.045]  # Параметры
tspan = (0.0, 180.0)                                         # Отрезок времени
# Задача
prob_pkpd = ODEProblem(pd_sde, pkpd_u0, tspan, pkpd_p)
# Решение, используется CVode Backward Differentiation Formula (BDF) solver из Sundials.jl
sol = solve(prob_pkpd, CVODE_BDF())
# График
plot(sol; idxs = [4,3], label = ["X1", "_"], xlims = (0, 120), ylims = (0, 20))
```

```julia
# Случайный шум

function σ_pkpdev(du, u, p, t)
    du[1] = u[1]*0.4
    du[2] = 0.0
    du[3] = 0.0
    du[4] = 0.0
    du[5] = 0.0
    du[6] = 0.0
    du[7] = 0.0
end

prob_sde_pkpdev = SDEProblem(pd_sde, σ_pkpdev, pkpd_u0, tspan, pkpd_p)
sde_sol = solve(prob_sde_pkpdev)
plot(sde_sol, idxs = [1, 4, 3], xlims=(0, 48), ylims = (0, 20))
```

```julia

ensembleprob_pkpd = EnsembleProblem(prob_sde_pkpdev)

ens_pkpd_sol = solve(ensembleprob_pkpd, EnsembleThreads(), trajectories = 1000)

summ_pkpd = EnsembleSummary(ens_pkpd_sol, 0:0.1:96)

plot(summ_pkpd, labels = "Middle 95%", idxs = [4, 3], xlims = (0, 96))

#summ = EnsembleSummary(ens_sol, 0:0.1:24; quantiles = [0.25, 0.75])
#plot!(summ, labels = "Middle 50%", legend = true)

```
